```python
# Группируем по snapshot_timestamp
result = [] # Создаём пустой список — сюда будем складывать словари (одна строка стакана = один словарь).

for ts, group in df.groupby('snapshot_timestamp'): #Проходим по группам строк, где ts — это уникальное значение snapshot_timestamp, а group — подтаблица с ордерами на этот момент времени.
    row = {'dt': ts} #Создаём новую строку будущего датафрейма (словарь), с ключом 'dt', равным текущему snapshot_timestamp.

    # Выделяем заявки на покупку (bid) и продажу (ask), и сортируем их по уровню, чтобы bid_1 был ближе к рыночной цене, bid_2 — второй по глубине и т.д.
    bids = group[group['type'] == 'bid'].sort_values('level')
    asks = group[group['type'] == 'ask'].sort_values('level')

    # Проходим по строкам с бид-заявками. i — это уровень (1, 2, 3...).
    for i, (_, bid) in enumerate(bids.iterrows(), 1): # iterrows() -> (index, row)
        row[f'bid_{i}'] = bid['price']
        row[f'size_bid_{i}'] = bid['size']

    # Проходим по строкам с аск-заявками. i — это уровень (1, 2, 3...).
    for i, (_, ask) in enumerate(asks.iterrows(), 1):
        row[f'ask_{i}'] = ask['price']
        row[f'size_ask_{i}'] = ask['size']
    
    result.append(row)

# Соберем всё в датафрейм
book_df = pd.DataFrame(result).sort_values('dt')
```

```python
def backtest_generic_with_baseline(
    df,
    strategy_fn,
    initial_cash=0,
    fee_rate=0.0002,         # комиссия (обычно 0,02% =0,0002)
    slippage=0.0002,           # проскальзывание, тоже в %
    verbose=False,
    **strategy_kwargs
):
    """
    Универсальный бэктестер + сравнение с Buy & Hold, с учетом комиссий и slippage.

    Аргументы:
        df: DataFrame с колонками как минимум ['dt', 'bid_1', 'ask_1']
        strategy_fn: функция стратегии, возвращает одну из ['buy', 'sell', 'hold']
        initial_cash: начальный баланс
        verbose: логировать ли действия
        **strategy_kwargs: любые параметры, которые хочет использовать стратегия
    """

    position = None
    entry_price = None
    cash = initial_cash
    pnl_history = []
    action_history = []
    entry_time = None

    buy_and_hold_entry_price = float(df.iloc[0]["ask_1"])
    buy_and_hold_pnl_history = []

    for i, row in df.iterrows():
        row = row.copy()
        dt = row["dt"]
        bid_raw = float(row["bid_1"])
        ask_raw = float(row["ask_1"])

        # Корректируем с учетом slippage
        ask = ask_raw + slippage
        bid = bid_raw - slippage

        history_df = df.iloc[:i].copy() if i > 0 else pd.DataFrame()

        action = strategy_fn(
            i=i,
            row=row,
            history=history_df,
            position=position,
            entry_price=entry_price,
            **strategy_kwargs
        )

        if action == "buy" and position is None:
            position = "long"
            entry_price = ask * (1 + fee_rate) * (1 + slippage)  # вход с комиссией и проскальзыванием
            entry_time = dt
            if verbose:
                print(f"{dt}: BUY at {ask:.2f} (incl. fee)")

        elif action == "sell" and position == "long":
            exit_price = bid * (1 - fee_rate) * (1 - slippage)  # выход с комиссией и проскальзыванием
            profit = exit_price - entry_price
            cash += profit
            if verbose:
                print(f"{dt}: SELL at {bid:.2f} (net: {exit_price:.2f}, PnL: {profit:.2f})")
            position = None
            entry_price = None
            entry_time = None

        unrealized = (bid * (1 - fee_rate) * (1 - slippage) - entry_price) if position == "long" else 0
        total_pnl = cash + unrealized
        pnl_history.append(total_pnl)
        action_history.append(action)

        # Buy & Hold тоже с учетом комиссий и slippage
        buy_and_hold_entry_net = buy_and_hold_entry_price * (1 + fee_rate) * (1 + slippage)
        buy_and_hold_exit_net = bid_raw * (1 - fee_rate) * (1 - slippage)
        buy_and_hold_pnl = buy_and_hold_exit_net - buy_and_hold_entry_net
        buy_and_hold_pnl_history.append(buy_and_hold_pnl)

    result = df.copy()
    result["pnl"] = pnl_history
    result["action"] = action_history
    result["buy_and_hold_pnl"] = buy_and_hold_pnl_history
    return result
```

```python
def strategy_price_drop_rise(i, row, history, position, entry_price, drop_pct=0.001, rise_pct=0.001):
    """
    Всё, что требуется — реализовать strategy_fn, 
    которая смотрит на 
    row, history, position, entry_price 
    и возвращает 'buy', 'sell' или 'hold'.
    """
    ask = float(row["ask_1"])
    bid = float(row["bid_1"])

    if position is None:
        if i == 0:
            return "hold"
            
        last_ask = float(history.iloc[-1]["ask_1"])
        if ask < last_ask * (1 - drop_pct):
            return "buy"
            
    elif position == "long":
        if bid > entry_price * (1 + rise_pct):
            return "sell"
            
    else:        
        return "hold"
```

```python
def strategy_mean_reversion(i, row, history, position, entry_price, lookback=20, entry_threshold=0.995, exit_threshold=1.005):
    """
    Стратегия Mean Reversion:
    - Покупает, если текущая ask_1 значительно ниже среднего ask за lookback-период.
    - Продаёт, если текущий bid_1 выше средней цены входа или средней цены за период.

    Аргументы:
        lookback: окно для расчета среднего ask
        entry_threshold: доля от среднего ask для входа (например, 0.995 = -0.5%)
        exit_threshold: доля от цены входа для выхода (например, 1.005 = +0.5%)
    """
    ask = float(row["ask_1"])
    bid = float(row["bid_1"])

    if i < lookback:
        return "hold"  # Не хватает истории

    recent_asks = history.iloc[-lookback:]["ask_1"].astype(float)
    mean_ask = recent_asks.mean()

    if position is None:
        if ask < mean_ask * entry_threshold:
            return "buy"

    elif position == "long":
        if bid > entry_price * exit_threshold:
            return "sell"

    return "hold"
```

```python
def plot_strategy_vs_baseline(result_df):
    plt.figure(figsize=(12, 6))
    plt.plot(result_df["dt"], result_df["pnl"], label="Strategy PnL", linewidth=2)
    plt.plot(result_df["dt"], result_df["buy_and_hold_pnl"], label="Buy & Hold PnL", linestyle="--", linewidth=2)
    plt.xlabel("Time")
    plt.ylabel("PnL")
    plt.title("Strategy vs Buy & Hold PnL")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# define metrics
def validation_date_range(df):
    return df["dt"].dt.floor("h").dt.strftime("%Y-%m-%d %H:%M").min(), df["dt"].dt.floor("h").dt.strftime("%Y-%m-%d %H:%M").max()

def number_of_trade_signals(df):
    return sum(df['action'].isin(['buy', 'sell']))

def total_deals_opened(df):
    return sum(df['action'] == 'buy')

def total_deals_closed(df):
    return sum(df['action'] == 'sell')

def winning_trade_percentage(df):
    wins = 0
    entry_price = None
    for _, row in df.iterrows():
        if row['action'] == 'buy':
            entry_price = row['ask_1']
        elif row['action'] == 'sell' and entry_price is not None:
            if row['bid_1'] > entry_price:
                wins += 1
    total_closed = total_deals_closed(df)
    return wins / total_closed * 100 if total_closed > 0 else 0

def final_pl(df):
    return round(df["pnl"].iloc[-1], 2)

def max_pl(df):
    return round(df["pnl"].max(), 2)

def min_pl(df):
    return round(df["pnl"].min(), 2)

def max_drawdown(df):
    cum_max = df["pnl"].cummax()
    drawdowns = cum_max - df["pnl"]
    return round(drawdowns.max(), 2)

def max_drawdown_period_range(df):
    df = df.copy()
    df["cum_max"] = df["pnl"].cummax()
    df["drawdown"] = df["cum_max"] - df["pnl"]

    # Индекс максимальной просадки
    max_drawdown_idx = df["drawdown"].idxmax()
    if pd.isna(max_drawdown_idx):
        return None

    # Индекс локального максимума перед началом просадки
    peak_idx = df.loc[:max_drawdown_idx]["pnl"].idxmax()

    # Дата начала и конца просадки
    start_time = df.loc[peak_idx, "dt"]
    end_time = df.loc[max_drawdown_idx, "dt"]

    # Возвращаем кортеж в виде строки
    return (start_time.strftime("%Y-%m-%d %H:%M"), end_time.strftime("%Y-%m-%d %H:%M"))


def positive_pl_hours(df):
    return round(sum(df["pnl"] > 0)/60, 1)

def turnover(df):
    return sum(df['action'].isin(['buy', 'sell']))

def turnover_cash(df):
    turnover = 0
    position = None

    for i, row in df.iterrows():
        action = row["action"]
        if action == "buy":
            price = float(row["ask_1"])
            turnover += price
            position = "long"
        elif action == "sell" and position == "long":
            price = float(row["bid_1"])
            turnover += price
            position = None

    return turnover
```

```python
def generate_metrics_report(df):
    metrics = {
        "Validation Range": [validation_date_range(df)],
        "Number of Trade Signals": [number_of_trade_signals(df)],
        "Total Deals Opened": [total_deals_opened(df)],
        "Total Deals Closed": [total_deals_closed(df)],
        "Winning Trade %": [winning_trade_percentage(df)],
        "Final PnL": [final_pl(df)],
        "Max PnL": [max_pl(df)],
        "Min PnL": [min_pl(df)],
        "Max Drawdown": [max_drawdown(df)],
        "Max Drawdown period": [max_drawdown_period_range(df)],
        "Positive PnL Hours": [positive_pl_hours(df)],
        "Turnover (Total Trades)": [turnover(df)],
        "Turnover (Cash)": [turnover_cash(df)],
    }
    return pd.DataFrame(metrics).T.rename(columns={0: "Value"})
```
